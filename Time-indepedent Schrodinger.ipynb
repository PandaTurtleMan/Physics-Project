```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.linalg import eigh_tridiagonal


import matplotlib.pyplot as plt
import numpy as np


#Note: everything is dimensionless, i.e h=1
```

```python
N = 2000 #splitting the line into n parts
dy = 1/N #If the line has unit length, then the size of each little interval is 1/n
y = np.linspace(0,1,N+1) #creates an array representing the discretized line, where index goes from 0 to n
```

```python
#this is where we define our (dimensionlesss) potential, 
#it can be basically whatever we want with any boundary conditions.
def mL2V(y):
    #return 1000*(1/(y+0.000000001)) #hydrogen, not quite working
    #return 1000*np.sin(20*y) * y**4 #weird function
    #return 1000*(y-1/2)**2 #quadratic/harmonic oscillator
    #return 1000*np.exp(-(y-1/2)**2) #normal distribution
    #return 1000*((y-1/2)**3)
    return 0*y

V = mL2V(y)
plt.plot(y,V)
```

```python
d = 1/dy**2 + mL2V(y)[1:-1]
e = -1/(2*dy**2) * np.ones(len(d)-1) #our system of equations will form a symmetric tridiagonal matrix; therefore
# we only need to define whats on the diagonal of the matrix and what's on the off diagonal

w, v = eigh_tridiagonal(d, e) #solves eigenvalues w and eigenvectors v
#for tri-diagonal matrix, might improve this bit later
```

```python
#Let's find the eigenvectors (functions)!
plt.plot(v.T[0])
plt.plot(v.T[1])
plt.plot(v.T[2])
plt.plot(v.T[3])
plt.plot(v.T[4])
#plt.plot(v.T[20]**2)
#notice for a potential of zero, we get the first five standing waves
```

```python
#Now the eigenvalues (energies)! Notice that they are evenly spaced with a minimum zero point
#energy, as expected for a harmonic oscillator
plt.bar(np.arange(0, 10, 1), w[0:10])
```
